```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Bengaluru_House_Data_Cleaned.csv")
df.head(3)
```

```python
# one hot coding 
dummies = pd.get_dummies(df.location)
dummies.head(3)
```

```python
df1 = pd.concat([df,dummies.drop('other', axis = 'columns')], axis = 'columns')
df1.head(3)
```

```python
df2 = df1.drop('location', axis = 'columns')
df2.head(2)
```

```python
df2.shape
```

```python
# X contains all independent variables
X = df2.drop('price',axis = 'columns')
X.head(3)
```

```python
# y contains dependent variable
y = df2.price
y.head(3)
```

```python
# split test 20% and train 80% 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 10)
```

```python
# linear regression model
from sklearn.linear_model import LinearRegression

# train model
l = LinearRegression()
l.fit(X_train, y_train)
```

```python
# tell good how your model is
l.score(X_test, y_test)
```

```python
df2.head(3)
```

```python
# predict function
def predict_price(location, total_sqft, bath, bhk):
    loc_index = np.where(X.columns == location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = total_sqft
    x[1] = bath
    x[2] = bhk
    
    if loc_index >=0:
        x[loc_index] = 1
    return l.predict([x])[0]
```

```python
predict_price('1st Block Jayanagar', 1000,2,2)
```

```python
predict_price('1st Block Jayanagar', 1000,3,2)
```
